```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('student-mat (1).csv')
print(df.head(5))
```

```python
print(df.isnull().sum())
```

```python
print(df.dtypes)
```

```python
df.shape
```

```python
missing_values = df.isnull().sum()
total_missing = missing_values.sum()
# Check for duplicate rows
duplicate_rows = df.duplicated().sum()
print(f"\nTotal Missing Values: {total_missing}")
print(f"Total Duplicate Rows: {duplicate_rows}")
```

```python

# 1. Calculate the average score in math (G3)
average_score_g3 = df['G3'].mean()
print(f"1. Average score in math (G3): {average_score_g3:.2f}")
```

```python
# 2. Count the number of students who scored above 15 in their final grade (G3)
students_above_15 = df[df['G3'] > 15].shape[0]
print(f"2. Number of students who scored above 15 in their final grade (G3): {students_above_15}")
```

```python
# 3. Calculate the correlation between study time and the final grade (G3)
correlation_studytime_g3 = df['studytime'].corr(df['G3'])
print(f"3. Correlation between study time and final grade (G3): {correlation_studytime_g3:.2f}")
```

```python
# 4. Calculate the average final grade (G3) for each gender
average_g3_by_gender = df.groupby('gender')['G3'].mean()
print("4. Average final grade (G3) by gender:")
print(average_g3_by_gender)
```

```python
# 1. Plot a histogram of final grades (G3)
plt.figure(figsize=(8, 5))
sns.histplot(df['G3'], bins=10, kde=True, color='blue', edgecolor='black')
plt.title('Histogram of Final Grades (G3)')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Create a scatter plot between study time and final grade (G3)
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['studytime'], y=df['G3'], color='green')
plt.title('Scatter Plot: Study Time vs Final Grade (G3)')
plt.xlabel('Study Time')
plt.ylabel('Final Grade (G3)')
plt.show()
```

```python
# 3. Create a bar chart comparing the average scores of male and female students
average_g3_by_gender = df.groupby('gender')['G3'].mean().reset_index()
plt.figure(figsize=(8, 5))
sns.barplot(x='gender', y='G3', data=average_g3_by_gender)
plt.title('Average Final Grades (G3) by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Final Grade (G3)')
plt.show()
```
